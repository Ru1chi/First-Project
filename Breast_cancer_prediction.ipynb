```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import sklearn.datasets
from sklearn.metrics import accuracy_score
```

Data Collection and Preporcessing

```python
#loading the data

data1 = sklearn.datasets.load_breast_cancer()
```

```python

print(data1)#its is in the form of numpy array
```

```python
#loading data into dataframe
dataframe=pd.DataFrame(data1.data,columns=data1.feature_names)
```

```python
dataframe.head()
```

```python
#ADDING THE TARGET COLUMN
dataframe['label']=data1.target
```

```python
dataframe.tail()
```

```python
#number of rows and columns in the dataset
dataframe.shape
```

```python
dataframe.info()
```

```python
#checking for missing values
dataframe.isnull().sum()
```

```python
#statistical measures of dataset
dataframe.describe()
```

```python
dataframe['label'].value_counts()
```

1=benign

0=malignant

```python
#group the data based on label
dataframe.groupby('label').mean()
```

```python
#seperating dependent and independent variables
x=dataframe.drop(columns='label',axis=1)
```

```python
x
```

```python
y=dataframe['label']
```

```python
y
```

```python
#train test split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Model
lr=LogisticRegression()
```

```python
#training the model using the training data
lr.fit(x_train,y_train)
```

```python
#Model Evaluation
x_train_prediction=lr.predict(x_train)
```

```python
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print("Accuracy if training data: ",training_data_accuracy)
```

```python
x_test_prediction=lr.predict(x_test)
testing_data_accuracy=accuracy_score(y_test,x_test_prediction)
print("Accuracy of testing data",testing_data_accuracy)
```
