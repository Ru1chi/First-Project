```python
#linear SVM-classify dataset in 2 classes by just drawing 1 line
#non-linear svm-cant classify by just drawing strainght line
#kernel function-takes low dimensional i/p and transform it into a higher-dimensional space(converts non-seperable problem into seperable problem)
import pandas as pd
```

```python
df=pd.read_csv("/content/iris.csv")
```

```python
df
```

```python
x=df.iloc[:,0:4]
```

```python
y=df.iloc[:,4]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#now import svm model
from sklearn.svm import SVC
```

```python
model=SVC(kernel = 'linear')
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
x_test
```

```python
model.score(x_test,y_test)
```
