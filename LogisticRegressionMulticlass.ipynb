```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("/content/iris.csv")
```

```python
df
```

```python
df["variety"].unique() #print all the dependent variables
```

```python
df["variety"].replace({'Setosa':'1', 'Versicolor':'2', 'Virginica':'3'},inplace=True)    #change the values of  categorical variables in to numbers
```

```python
df
```

```python
#now we need to split the data
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['sepal.length',	'sepal.width',	'petal.length', 'petal.width']],df['variety'],test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
#now we need to train the model by applyrin fittin
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
lr.predict(y_test)#to test the model it requires a 2D array and y_test is not 2D array
```

```python
#Now we can check the score of the model
```

```python
lr.score(x_test,y_test)
```
